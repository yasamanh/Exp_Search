```python
import json
```

```python
f = open("/Users/alinamazi/Data/results/April/Followers/ua_account_march_april.json",'r')
data = json.load(f)
tweet_data = {}
tweet_data["tweets"] = data['tweets']
f.close()
```

```python
count = 0
followers_list = []
fav_list = []
fr_list = []
time_list = []
screen_name = []
for d in tweet_data['tweets']:
    if 'user' in d:
            count += 1
            fr_list.append(d['user']['friends_count'])
            fav_list.append(d['user']['favourites_count'])
            followers_list.append(d['user']['followers_count'])
            time_list.append(d['created_at'])

print ("%s %s" %( len(time_list), len(followers_list)))
```

```python
from datetime import datetime
new_time = []
for t in time_list:
    ts = datetime.strptime(t,'%a %b %d %H:%M:%S +0000 %Y')
    new_time.append(ts)
```
