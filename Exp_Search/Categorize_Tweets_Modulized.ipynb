```python
import os
import time
import gzip
import json as simplejson
import pprint
import pdb
```

```python
def Query_Tweets(tweets_path, results_path):

    json_count = 0
    tweet_count = 0
    eng_count = 0
    
    bbc = []
    alj = []
    nytimes = []
    reuters = []
    yahoonews = []
    breakingnews = []
    npr = []
    cnn = []
    addele = []
    nasa = []
    food = []
    election = []
    politics = []
    entertainment = []
    wellness = []
    sport = []
    sport_media = []
    football = []
    espn = []

    t0 = time.time()
    for file in os.listdir(tweets_path):
        if file.endswith(".gz"):
            f=gzip.open(tweets_path+file,'rb')
            print "start reading file: ", file
            try:
                t1= time.time()
                for l in f:
                    #l = l.decode("ascii","ignore").encode("utf-8")
                    l = l.decode("ascii","ignore").encode("ascii")
                    data = simplejson.loads(l)
                    #print(data)

                    if "lang" in data:
                        if data["lang"] == 'en' and "text" in data:
                            eng_count += 1

                            text = data["text"].encode('utf-8').lower().replace('\n',' ')
                            text_id = str(data["id"])+' '+text
                            source_name = source_detection(text)
                            if source_name: 
                                eval(source_name).append(text_id)

                    if "text" in data and "created_at" in data and "id_str" in data:
                        tweet_count += 1

                    json_count += 1
                f.close()
                t2 = time.time()
                print "done with file: ", file, ' This file takes {} seconds'.format(t2-t1)
        
            except ValueError:
                print ValueError
                pass
            
    t3 = time.time()
    print 'Going through all the files takes {} seconds'.format(t3 - t0)

    print "Total number of JSON: ", json_count
    print "Total number of valid tweets:", tweet_count
    print "Total number of english tweets", eng_count
    
    #merging all newswire sources together
    newswire_sources = reuters + nytimes + yahoonews + alj + bbc + npr + cnn + breakingnews
    print "Number of newswire tweets ", len(newswire_sources)
    
    print "Number of @bbc ", len(bbc)
    print "Number of @cnn ", len(cnn)
    print "Number of @nyt ", len(nytimes)
    print "Number of @npr ", len(npr)
    print "Number of @breakingnews ", len(breakingnews)
    print "Number of @reuters ", len(reuters)
    print "Number of @yahoonews ", len(yahoonews)
    print "Number of @alj ", len(alj)
    
    print 
    print "Number of @addele ", len(addele)
    print "Number of @nasa ", len(nasa)
    print "Number of @food ", len(food)
    print "Number of @election ", len(election)
    print "Number of @politics ", len(politics)
    print "Number of @entertainment", len(entertainment)
    print "Number of @wellness ", len(wellness)
    print "Number of @SPORTMEDIA ", len(sport_media)
    print "Number of @SPORT ", len(sport)
    print "Number of @ESPN ", len(espn)
    print "Number of @football ", len(football)
        
   ##Writing the results file
    results_to_file("nw", results_path, newswire_sources)
    results_to_file("food", results_path, food)
    results_to_file("election", results_path, election)
    results_to_file("politics", results_path, politics)
    results_to_file("entertainment", results_path, entertainment)
    results_to_file("wellness", results_path, wellness)
    results_to_file("nasa", results_path, nasa)
    results_to_file("addele", results_path, addele)
    results_to_file("football", results_path, football)
    results_to_file("espn", results_path, espn)
    results_to_file("sport", results_path, sport)
    results_to_file("sport_media", results_path, sport_media)
    
    
```

```python

def results_to_file(category, results_path, array):
    switcher = {
        "nw": "nw.txt",
        "food": "food.txt",
        "politics": "politics.txt",
        "wellness": "wellness.txt",
        "entertainment": "entertainment.txt",
        "election": "election.txt",
        "nasa":"nasa.txt",
        "addele" : "addele.txt",
        "sport_media": "sport_media.txt",
        "sport": "sport.txt",
        "espn": "espn.txt",
        "football": "football.txt"
        }
    file_name = switcher.get(category)
    o_f = open(results_path + file_name, 'w')
    for item in array:
        o_f.write("%s\n" % item.decode("ascii","ignore").encode('utf-8'))
        #o_f.write("%s\n" % item.encode('utf-8'))
    o_f.close()
```

```python
def news_selection(argument):
    switcher = {
                "cnn": "cnn_a",
                "cnnbrk": "cnn_a",
                "ajam": "alj_a",
                "ajenglish": "alj_a",
                "yahoonews": "yahoonews_a",
                "reuters": "reuters_a",
                "reutersopinion": "reuters_a",
                "breakingnews": "breakingnews_a",
                }
    
    f = switcher.get(argument, "nothing")
    return f
```

```python
#example of how to handle switch and funcation calls
def zero():
    return "zero"

def one():
    return "one"

def numbers_to_functions_to_strings(argument):
    switcher = {
        0: zero,
        1: one,
        2: lambda: "two",
    }
     # Get the function from switcher dictionary
    func = switcher.get(argument, lambda: "nothing")
    # Execute the function
    return func()
```
