```python
import os
import time
import gzip
import json
import pprint
import pdb
```

```python
def Query_Tweets(tweets_path, results_path):

    json_count = 0
    tweet_count = 0
    eng_count = 0
    
    tweets_data = []

    uber = []

    delete = []

    import os.path
    if not os.path.isdir(tweets_path):
        print 'wrong path'
        return
    if not os.path.isdir(results_path):
        print 'wrong ptath'
        return
    
    
    json_file_uber = open(results_path + 'uber.json', 'w')
    
    t0 = time.time()
    for file in os.listdir(tweets_path):
        if file.endswith(".gz"):
            f=gzip.open(tweets_path+file,'r')
            print("start reading file: ", file)
            try:
                t1= time.time()
                for l in f:
                    l = l.decode("ascii","ignore").encode("ascii")
                    data = json.loads(l)
                    
                    if "lang" in data:
                        if data["lang"] == 'en' and "text" in data:
                            eng_count += 1

                            text = data["text"].encode('utf-8').lower().replace('\n',' ')
                            #text_id = str(data["id"])+' '+text
                            source_name = source_detection(text)
                            if source_name:
                                eval(source_name).append(data)
                                #eval(source_name).append((text_id))
                                
                                #tweets_data.append(data)


                    if "text" in data and "created_at" in data and "id_str" in data:
                        tweet_count += 1

                f.close()
                t2 = time.time()
                print("done with file: ", file, ' This file takes {} seconds'.format(t2-t1))
        
            except ValueError:
                print(ValueError)
                pass
            
    t3 = time.time()
    print('Going through all the files takes {} seconds'.format(t3 - t0))

    print("Total number of valid tweets:", tweet_count)
    print("Total number of english tweets", eng_count)
    print("Total number of JSON of selected tweets: ", json_count)
    
    #merging all newswire sources together
    uber = uber + delete 
    print("Number of Uber related tweets ", len(uber))
    
    ##Print break down number of tweets
    print("Number of @uber ", len(uber))
    print("Number of #deleteuber", len(delete))
    
    ##Writing the results file
    #results_to_file("ua", results_path, ua)
    #results_to_file("aa", results_path, aa)
    
         
    def jdefault(o):
        return o.__dict__
    
    
    tweet_dic_uber = {}

    tweet_dic_uber['tweets'] = uber
    json_data_uber = json.dumps(tweet_dic_uber, default=jdefault)
    json_file_uber.writelines(json_data_uber)
    json_file_uber.close()
    
```

```python

def results_to_file(category, results_path, array):
    switcher = {
        "uber": "uber.txt",
        "aa": "aa.txt",
        }
    file_name = switcher.get(category)
    o_f = open(results_path + file_name, 'w')
    for item in array:
        o_f.write("%s\n" % item.decode("ascii","ignore").encode('utf-8'))
        #o_f.write("%s\n" % item.encode('utf-8'))
    o_f.close()
```

```python
def source_detection(ttext):
    source = ""

    if "@uber " in ttext:
        source = "uber"
    
    if "#uber " in ttext:
        source = "uber"
    
    if "#deleteuber" in ttext:
        source = "uber"

    return source
```
